```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
f2 = "sim_LOS_modifier_200.txt"
sim_ds = []
sim_pathloss = []
sim_ds2 = []
sim_ave_path = []
sim_difference = []
sim_corr = []
with open(f2) as csvfile:
    rc = csv.reader(csvfile)
    first = True
    for row in rc:
        if first:
            first = False
            continue
        sim_ds.append(float(row[0]))
        sim_pathloss.append(-1*float(row[1]))
        sim_ds2.append(float(row[2]))
        sim_ave_path.append(float(row[3]))
        diff = float(row[2])-float(row[3])
        corr = 1.0/532 * float(row[2])
        sim_difference.append(diff)
        sim_corr.append(diff-corr)
```

```python
plt.rcParams["figure.figsize"] = (6,4)
color_RT = '#5d3a9b'#'#1aff1a'#'#006cd1'
plt.plot(sim_ds, sim_difference, label="Ave_path_length", color=color_RT)
plt.grid(True)
plt.show()
```

```python
plt.plot(sim_ds, sim_corr, label="Ave_path_length", color=color_RT)
plt.grid(True)
plt.show()
```

```python
print(sum(sim_corr))
```
