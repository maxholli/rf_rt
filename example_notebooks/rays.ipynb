# Finding the reflection coefficient
for vertically polarized waves

```python
c = 300000000.0
f = 900000000.0
d = 500.0
h_T = 50
h_R = 2.5
d_1 = np.sqrt(np.power(h_T-h_R, 2) + np.power(d, 2)) # distance from h_T to h_R
## finding psi, the incident angle 
crow_d = (-1*h_T) / ((-1 * (h_T + h_R))/d) # distance from h_T to reflection point
psi_rad = np.arctan(h_T/crow_d)
psi_deg = 360*psi_rad/(2*np.pi)
print("psi {:.5f} radians or {:.5f} degrees ".format(psi_rad, psi_deg))

## calculating the reflection coefficient for vertically polarized wave
epsilon = 15                 # dielectric constant of the average ground
sigma = 12e-3                 # conductivity of average ground
epsilon_fs = 8.8541878128e-12      # dielectric constant of free space aka vacuum permitivity 
omega = 2 * np.pi * f        # angular frequency of transmission
print("omega", omega)
x = sigma / (omega * epsilon_fs)
x_check = 18e9 * sigma / f
print("the following two numbers should be the same\n{:.5f} {}".format(x, x_check))
## finally calculating the reflection coefficient (vertically polarized)

comp_part = epsilon - 1j*x
rnum = comp_part * np.sin(psi_rad) - np.sqrt(comp_part - np.power(np.cos(psi_rad),2))
rden = comp_part * np.sin(psi_rad) + np.sqrt(comp_part - np.power(np.cos(psi_rad),2))
rho_v = rnum / rden
print("reflection coefficient {:.10f} + {:.10f}j".format(rho_v.real, rho_v.imag))
mag = np.sqrt(np.power(rho_v.real,2)+ np.power(rho_v.imag,2))
print("magnitude {:.4f} and phase {:.2f} degrees".format(
    mag,
    -1*np.arccos(rho_v.real / mag)*360 / (2*np.pi)
    ))
```

# Finding 2-ray field strength
For distances less than a few tens of kilometers, it is ofter permissible to neglect the Earth curvature, which is exactly what we'll do.

```python
## find phase difference between the direct ray and the reflected ray
d_2 = np.sqrt(np.power(h_T+h_R, 2) + np.power(d, 2))                    # distance traveled by reflected ray
delta_d = d_2 - d_1                                                       # difference in distance between direct and reflected rays
# delta_d = 502.775 - 502.283
print("direct len {:.3f} reflected len {:.3f}".format(d_1, d_2))
lamb = c/f                                                              # wavelength meters
print("lambda {:.5f}".format(lamb))
delta_phi = (2*np.pi/lamb) * delta_d                                    # phase difference between direct and reflected rays
# delta_phi = delta_phi - 2*np.pi
print("phase diff {:.5f} radians, {:.5f} degrees".format(
    delta_phi,
    delta_phi*360 / (2*np.pi)
    ))
ce_phi = np.cos(delta_phi) - (1j * np.sin(delta_phi))
print("ce_phi is ", ce_phi)
print("magnitude {:.5f} contribution from phase difference".format(
    np.sqrt(np.power(ce_phi.real,2)+ np.power(ce_phi.imag,2))
))
change_E = 1 + rho_v * ce_phi
print(change_E)
```

```python
print("{:.10f} + {:.10f}j".format(change_E.real, change_E.imag))
print("change in E, magnitude {:.1f} and phase {:.5f} degrees".format(
    np.sqrt(np.power(change_E.real,2)+ np.power(change_E.imag,2)),
    np.arctan(change_E.imag / change_E.real)*360 / (2*np.pi)
    ))
```

```python
## direct ray 
direct_ray_power = np.power(lamb/(4*np.pi*d), 2)
print("Power of the direct ray = {:.15f}".format(direct_ray_power))
reflect_ray_power = np.power(change_E.real,2)+ np.power(change_E.imag,2)
print("Power change from reflected = {:.15f}".format(reflect_ray_power))

ref_ray_check = 4*np.power(np.sin((2*np.pi*h_T*h_R)/(lamb*d_1)),2)
print("check: Power change from reflected = {:.15f}".format(ref_ray_check))
```

```python
two_ray_prop_gain = direct_ray_power * reflect_ray_power
two_ray_check = direct_ray_power * ref_ray_check
print("{:.15f}\n{:.15f}".format(two_ray_prop_gain, two_ray_check))
print("2-ray prop gain = {:.2f}".format(10*np.log10(two_ray_prop_gain)))
print("check: 2-ray prop gain = {:.2f}".format(10*np.log10(two_ray_check)))
print("direct-ray gain = {:.2f} (aka fspl)".format(10*np.log10(direct_ray_power)))
```

## recreate fig 2.5 from the book

```python
ds = np.linspace(5,1000,1000)

c = 300000000.0
f = 900000000.0
# d = 10000.0
h_T = 50
h_R = 5
d_1 = np.sqrt(np.power(h_T-h_R, 2) + np.power(ds, 2)) # distance from h_T to h_R
## finding psi, the incident angle 
crow_d = (-1*h_T) / ((-1 * (h_T + h_R))/ds) # distance from h_T to reflection point
psi_rad = np.arctan(h_T/crow_d)
psi_deg = 360*psi_rad/(2*np.pi)
# for i in range(len(psi_rad)):
#     print("psi {:.5f} radians or {:.5f} degrees ".format(psi_rad[i], psi_deg[i]))

## calculating the reflection coefficient for vertically polarized wave
epsilon = 15                 # dielectric constant of the average ground
sigma = 12e-3                 # conductivity of average ground
epsilon_fs = 8.8541878128e-12      # dielectric constant of free space aka vacuum permitivity 
omega = 2 * np.pi * f        # angular frequency of transmission
x = sigma / (omega * epsilon_fs)
# x_check = 18e9 * sigma / f
# print("the following two numbers should be the same\n{:.5f} {}".format(x, x_check))
## finally calculating the reflection coefficient (vertically polarized)

comp_part = epsilon - 1j*x
rnum = comp_part * np.sin(psi_rad) - np.sqrt(comp_part - np.power(np.cos(psi_rad),2))
rden = comp_part * np.sin(psi_rad) + np.sqrt(comp_part - np.power(np.cos(psi_rad),2))
rho_v = rnum / rden
# print("reflection coefficient {:.10f} + {:.10f}j".format(rho_v.real, rho_v.imag))
mag = np.sqrt(np.power(rho_v.real,2)+ np.power(rho_v.imag,2))

## find phase difference between the direct ray and the reflected ray
d_2 = np.sqrt(np.power(h_T+h_R, 2) + np.power(ds, 2))                    # distance traveled by reflected ray
delta_d = d_2 - d_1                                                       # difference in distance between direct and reflected rays
# print("direct len {:.1f} reflected len {:.1f}".format(d_1, d_2))
lamb = c/f                                                              # wavelength meters
# print("lambda {:.5f}".format(lamb))
delta_phi = (2*np.pi/lamb) * delta_d                                    # phase difference between direct and reflected rays
# print("phase diff {:.5f} radians, {:.5f} degrees".format(
#     delta_phi,
#     delta_phi*360 / (2*np.pi)
#     ))

ce_phi = np.cos(delta_phi) - (1j * np.sin(delta_phi))
change_E = 1 + rho_v * ce_phi

direct_ray_power = np.power(lamb/(4*np.pi*d_1), 2)
reflect_ray_power = np.power(change_E.real,2)+ np.power(change_E.imag,2)

plt.plot(ds, 10*np.log10(reflect_ray_power))
plt.show()
```

```python
print("two ray path loss, over distance")
tot_power = 10*np.log10(direct_ray_power * reflect_ray_power)
plt.plot(ds, tot_power)
plt.show()
```

## Using the "Three-Ray" Model

```python
def find_reflection_coef(psi_rad, epsilon, sigma, f, polarity):
    c = 300000000.0
    epsilon_fs = 8.8541878128e-12      # dielectric constant of free space aka vacuum permitivity
    omega = 2 * np.pi * f        # angular frequency of transmission
    x = sigma / (omega * epsilon_fs)
    if polarity == "vertical":
        comp_part = epsilon - 1j*x
        rnum = comp_part * np.sin(psi_rad) - np.sqrt(comp_part - np.power(np.cos(psi_rad),2))
        rden = comp_part * np.sin(psi_rad) + np.sqrt(comp_part - np.power(np.cos(psi_rad),2))
        rho_v = rnum / rden
        return rho_v
    elif polarity == "horizontal":
        comp_part = epsilon - 1j*x
        rnum = np.sin(psi_rad) - np.sqrt(comp_part - np.power(np.cos(psi_rad),2))
        rden = np.sin(psi_rad) + np.sqrt(comp_part - np.power(np.cos(psi_rad),2))
        rho_v = rnum / rden
        return rho_v
    else:
        print("ERROR we dont have that polarization yet.")
```

```python
c = 300000000.0
## calculating the reflection coefficient for vertically polarized wave
epsilon = 15                 # dielectric constant of the average ground
sigma = 12e-3                 # conductivity of average ground
f = 1000000.0
psi_rad = 1.0472
rho_v1 = find_reflection_coef(psi_rad, epsilon, sigma, f, "vertical")

## path 2
psi_rad = 1.0472
rho_v2 = find_reflection_coef(psi_rad, epsilon, sigma, f, "horizontal")

print("1 reflection coefficient {:.5f} + {:.5f}j".format(rho_v1.real, rho_v1.imag))
print("2 reflection coefficient {:.5f} + {:.5f}j".format(rho_v2.real, rho_v2.imag))
mag = np.sqrt(np.power(rho_v1.real,2)+ np.power(rho_v1.imag,2))
print("magnitude {:.4f} and phase {:.2f} degrees".format(
    mag,
    -1*np.arccos(rho_v1.real / mag)*360 / (2*np.pi)
    ))


# delta_d = 0.001         # difference in distance between direct and reflected rays
# lamb = c/f              # wavelength meters
# delta_phi = (2*np.pi/lamb) * delta_d         # phase difference between direct and reflected rays
# ce_phi = np.cos(delta_phi) - (1j * np.sin(delta_phi))
# change_E = 1 + rho_v * ce_phi

# direct_ray_power = np.power(lamb/(4*np.pi*500), 2)
# reflect_ray_power = np.power(change_E.real,2)+ np.power(change_E.imag,2)

# two_ray_prop_gain = direct_ray_power * reflect_ray_power
# print("2-ray prop gain = {:.2f}".format(10*np.log10(two_ray_prop_gain)))
# # print("direct-ray gain = {:.2f} (aka fspl)".format(10*np.log10(direct_ray_power)))
```
